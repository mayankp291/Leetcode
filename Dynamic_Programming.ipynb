# DP vs Greedy

**Greedy** : Follow a predefined procedure to get optimal results. Make decision only once.

**DP** : List out all possible solutions and choose the best one. Make decisions at every step.



```python
# recursion for fibonacci series, same value calculated multiple times
# Time O(2^n)
def fib (n):
  if n <= 1:
    return n
  return fib(n-1) + fib(n-2)
```

**Memoization** : Using an array to store calculated values to avoid redundant work

```python
# memoization for fibonacci series, values are stored in array
# top down approach (tree is generated from top to down)
# Time O(n)

def fib (n):
  if F[n] != -1:
    return F[n]
  if n <= 1:
    return n
  F[n] = fib(n-1) + fib(n-2)
  return F[n]

n = 6
F = [-1] * (n+1)
print(fib(6))
```

**Tabulation** : Using an array to store calculated values to avoid redundant work, usually use an iterative method

```python
# tabulation for fibonacci series, values are stored in array
# bottom up approach, array is filled from left to right (Smaller value first) 
# iterative approach
# Time O(n)

def fib (n):
  if n <= 1:
    return n
  F[0], F[1] = 0, 1
  for i in range(2, n+1):
    F[i] = fib(i-1) + fib(i-2)
  return F[n]

n = 6
F = [-1] * (n+1)
print(fib(6))
```

# Dynamic Programming

*   Used to solve optimisation problems
*   Sequence of decisions






```python
# GRID TRAVELLER MEMOIZATION
# ways to travel between 2 points in a grid with only down and right allowed

# Time O(mn), Space O(m+n)

# [rows, cols]
# Base cases: [0,x]/[x,0] = 0, [1,1] = 1

def gridTraveler(m,n, memo={}):
  if (m,n) in  memo:
    return memo[(m,n)]
  if m == 1 and n == 1:
    return 1
  if m == 0 or n == 0:
    return 0
  # going down + going right
  memo[(m,n)] = gridTraveler(m-1, n) + gridTraveler(m, n-1)
  return memo[(m,n)]

print(gridTraveler(1,1))  #1
print(gridTraveler(2,3))  #3
print(gridTraveler(3,2))  #3
print(gridTraveler(3,3))  #6
print(gridTraveler(18,18))  #2333606220

```
